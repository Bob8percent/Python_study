```python
df = pd.DataFrame(data=[[1, 2, 3, np.nan, 4], 
                        [5, np.nan, 6, np.nan, 7],
                        [8, 9, 10, np.nan, 11],
                        [12, np.nan, np.nan, np.nan, 13],
                        [14, 15, 16, 17, 18]],
                  index=['A', 'B', 'C', 'D', 'E'],
                  columns=['C1', 'C2', 'C3', 'C4', 'C5'])
#   欠損値の処理
df.dropna() #   欠損値のない行を抽出
df['C2'].dropna()   #   C2のうち欠損値ではない部分を抽出
df[df['C2'].isnull()==False]    #   C2行の欠損してない行を抽出
df.dropna(thresh=3) #   3つ欠損している行は消す
df.dropna(thresh=3,axis=1)  #   3つ欠損している列は消す

df['C2'].fillna(df['C2'].mean())    #   行C2の欠損値をC2の平均値で埋める
```

```python
df = pd.DataFrame({'C1': ['A', 'A', 'A', 'B', 'B', 'C', np.nan],
                   'C2': [20, 50, 60, 80, 100, 30, 50],
                   'C3': [40, 200, 100, 500, 40, 200, 40]}) #   (書き方が新しいだけ)
df['C1'].value_counts()
df['C1'] = df['C1'].fillna(df['C1'].mode()[0]) #   行C1の欠損値を最頻値で埋める(更新するためにdf['C1']に代入(inplaceと同様))
round(df['C1'].value_counts()/len(df),2)

df.groupby('C1').sum()  #   行C1でまとめて合計値をだす
df.groupby('C1').max()
```

```python
#   関数の適用
df = pd.DataFrame(data=np.random.randn(5, 5), index=['A', 'B', 'C', 'D', 'E'],
                 columns=['C1', 'C2', 'C3', 'C4', 'C5'])

#   引数が一つのとき
def square(x):
    return x*x
df['C1'].apply(square)  #   行C1の値にsquareを適用できる

#   引数が複数のとき
def add(x, y):
    return x+y
add(df['C1'], df['C2'])

#   返り値が複数のとき
def square_and_cube(x):
    return pd.Series([x**2, x**3])  #   pd.Seriesに返り値を入れる
df[["squared", "cube"]]=df['C1'].apply(square_and_cube)
df
```
