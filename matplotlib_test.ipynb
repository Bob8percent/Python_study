```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
x=np.linspace(0,1,11)
y=2*x+0.1*np.random.randn(11)
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,1,11)
y=2*x+0.1*np.random.randn(11)
fig=plt.figure()    #   描画する環境を指定
ax=fig.add_subplot(111) #   環境中の描画場所を指定(111:1行1列の1番目に書く)

plt.plot(x,y)
plt.show()
```

```python
x_1 = np.linspace(0, 1, 10)
y_1 = x_1 + 0.1 * np.random.randn(10)

x_2 = np.linspace(0, 1, 100)
y_2 = 2 * x_2 + 0.1 * np.random.randn(100)

x_3 = np.linspace(0, 1, 1000)
y_3 = 3 * x_3 + 0.1 * np.random.randn(1000)

fig=plt.figure(figsize=(20,12))
ax_1=fig.add_subplot(221)
ax_2=fig.add_subplot(222)
ax_3=fig.add_subplot(223)

ax_1.plot(x_1, y_1)
ax_2.plot(x_2, y_2)
ax_3.plot(x_3, y_3)

plt.show()
```

```python
#fig, axを同時にとる
fig, axes =plt.subplots(2,2,figsize=(20,12))
axes.ravel()[0].plot(x_1,y_1) #  1次元配列に変換
axes.ravel()[1].plot(x_2,y_2)
axes.ravel()[2].plot(x_3,y_3)

plt.show()
```

```python
#   折れ線はplotメソッドを使う
df=pd.DataFrame(data=np.random.rand(30,2),columns=['C1','C2'])
fig, axis=plt.subplots(1,1,figsize=(7,5))
axis.plot(df['C1'],linewidth=5.0,linestyle="solid")
axis.plot(df['C2'],linewidth=5.0,linestyle="dashed")
axis.set_title("test",fontsize=20)
axis.set_xlabel("x",fontsize=15)
axis.set_ylabel("y",fontsize=15)
axis.legend(["data1","data2"])
plt.show()
```

```python
#   散布図はscatterメソッドを使う
df=pd.DataFrame(data=np.random.rand(500,2),columns=['C1','C2'])
fig, axis=plt.subplots(1,1,figsize=(7,5))
axis.scatter(df['C1'],df['C2'],alpha=0.5)
axis.set_title("test",fontsize=20)
axis.set_xlabel("x",fontsize=15)
axis.set_ylabel("y",fontsize=15)
plt.show()
```

```python
df=pd.DataFrame(data=np.random.rand(500,2),columns=['C1','C2'])
df["datasize"]=np.random.randint(0,100,500)
fig, axis=plt.subplots(1,1,figsize=(7,5))
axis.scatter(df['C1'],df['C2'],s=df['datasize']*10,c=df['datasize'],alpha=0.5)
axis.set_title("test",fontsize=20)
axis.set_xlabel("x",fontsize=15)
axis.set_ylabel("y",fontsize=15)
plt.show()
```

```python
#   棒グラフはbarメソッドを使う:棒グラフが扱うデータはカテコライズされている場合が多い
df = pd.DataFrame({'C1': ['A', 'A', 'A', 'B', 'B', 'C'],
                   'C2': [20, 50, 60, 80, 100, 30],
                   'C3': [40, 200, 100, 500, 40, 200,]})
fig, axis=plt.subplots(1,1,figsize=(7,5))
axis.bar(df['C1'].value_counts().index,df.groupby('C1').sum()['C2'])  
plt.show()
```

```python
#   複数の棒グラフについて
df = pd.DataFrame(data=100 * np.random.rand(5, 5), 
                  index=['A', 'B', 'C', 'D', 'E'], 
                  columns=['C1', 'C2', 'C3', 'C4', 'C5'])
fig, axis=plt.subplots(1,1,figsize=(7,5))
x=np.arange(len(df))
bar_width=0.4
axis.bar(x, df['C1'],width=bar_width,color="red",label='C1')#   引数：座標(x, y)
axis.bar(x+bar_width,df['C2'],width=bar_width,color="blue",label='C2')#   引数：座標(x, y)

plt.xticks(x+bar_width/2,df.index,fontsize=15)  #   labelを扱う関数(引数：位置、labelにしたい文字、...) 
plt.legend()
plt.show()
```

```python
#   練習
df = pd.DataFrame(data=100 * np.random.rand(5, 5), 
                  index=['A', 'B', 'C', 'D', 'E'])
fig,axis=plt.subplots(1,1,figsize=(7,5))
bar_width=0.1
x=np.arange(len(df),dtype='float64')
for i in df.columns:
    axis.bar(x,df[i],width=bar_width)
    x+=bar_width

axis.legend(df.index)
plt.xticks(np.arange(5)+bar_width*0.5*len(df),df.index,fontsize=15)
plt.show()
```

```python
#   ヒストグラムにはhistメソッドを使う
df = pd.DataFrame(data=np.random.randn(500), columns=['C1'])
fig,axis=plt.subplots(1,1,figsize=(7,5))
axis.hist(df['C1'],bins=20,rwidth=0.8)
plt.show()
```

```python
#   円グラフにはpieメソッドを使う
np.random.seed(0)
df = pd.DataFrame(data=np.random.randint(0, 5, 500), columns=['C1'])
fig,axis=plt.subplots(1,1,figsize=(7,5))
axis.pie(df['C1'].value_counts(),       #   引数：データの大きさ、
         labels=df['C1'].value_counts().index,
         autopct='%.1f%%',
         startangle=90,
         counterclock=False    #   時計回り
         ) 

plt.show()
```

```python
#   箱ひげ図はboxplotメソッド
df = pd.DataFrame(data=np.random.randint(0, 100, 500), columns=['C1'])
fig,axis=plt.subplots(1,1,figsize=(7,5))
axis.boxplot(df['C1'])
```

```python
#   グラフを画像ファイルとして出力したいときはsavefileメソッド
x = np.linspace(0, 1, 10)
y = 2 * x + np.random.randn(10)
fig,ax=plt.subplots(1,1,figsize=(7,5))
ax.plot(x,y)
plt.show()
plt.savefig('test.png')
```
